```python
# Importing Libraries
from bs4 import BeautifulSoup

import urllib.request

import nltk

from nltk.corpus import stopwords

from wordcloud import WordCloud

import matplotlib.pyplot as plt
```

```python
# connecting to website and fetching the response

response = urllib.request.urlopen('http://php.net/')

html = response.read()

print(html)
```

```python
# using beautifulsoup to clean the text
soup = BeautifulSoup(html,"html5lib")

text = soup.get_text(strip=True)

print(text)
```

```python
# converting text to tokens
tokens = [t for t in text.split()]

print(tokens)
```

```python
# counting word frequency
freq = nltk.FreqDist(tokens)

for key,val in freq.items():

    print (str(key) + ':' + str(val))
```

```python
#plotting words distribution

freq.plot(20, cumulative = False)
```

# Coding Exercise

## 1. With Stop Words

```python
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')

html = response.read()

print(html)
```

```python
# using beautifulsoup to clean the text
soup = BeautifulSoup(html,"html5lib")

text = soup.get_text(strip=True)

print(text)
```

```python
# converting text to tokens
tokens = [t for t in text.split()]

print(tokens)
```

```python
# counting word frequency
freq = nltk.FreqDist(tokens)

for key,val in freq.items():

    print (str(key) + ':' + str(val))
```

```python
#plotting words distribution

freq.plot(10, cumulative = False)
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10,max_words = 10).generate_from_frequencies(freq)

# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```

## 2. Without Stop Words and less frequency words

```python
# cleaning tokens and removing stop words
clean_tokens = tokens[:]

for token in tokens:

    if token in stopwords.words('english'):

        clean_tokens.remove(token)

freq = nltk.FreqDist(clean_tokens)

# for key,val in freq.items():

#     print (str(key) + ':' + str(val))
```

```python
# getting words which occured less than 5 times
freq_lt_5 = []
freq_gt_5 = []

for key,val in freq.items():
    if val > 5:
        freq_gt_5.append(key)
    else:
        freq_lt_5.append(key)   
```

```python
# removing words which occured less than 5 times
for less_freq_words in freq_lt_5:
    clean_tokens.remove(less_freq_words)
```

```python
# counting the frequency
freq_final = nltk.FreqDist(clean_tokens)
```

```python
#plotting words distribution

freq_final.plot(10, cumulative = False)
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10,max_words = 10).generate_from_frequencies(freq_final)

# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
